# Comparación de matrices de distancias y dendrogramas

\
Se describen procedimientos para comparar los resultados de distintas matrices de distancia. Comparación de árboles.
Se cargan 6 distintas matrices de distancias obtenidas de distintos subsets de marcadores y que fueron calculadas en DARwin version 2.3.4 usando el coeficiente de Jaccard.

```r
matriz1_dist_jaccard_darwin <- readRDS("data/matriz1_jaccard_Darwin.Rds")
matriz2_dist_jaccard_darwin <- readRDS("data/matriz2_jaccard_Darwin.Rds")
matriz3_dist_jaccard_darwin <- readRDS("data/matriz3_jaccard_Darwin.Rds")
matriz4_dist_jaccard_darwin <- readRDS("data/matriz4_jaccard_Darwin.Rds")
matriz5_dist_jaccard_darwin <- readRDS("data/matriz5_jaccard_Darwin.Rds")
matriz6_dist_jaccard_darwin <- readRDS("data/matriz6_jaccard_Darwin.Rds")
```

\
Para las comparaciones se verifica que todas las matrices tengan el mismo orden en filas y columnas. Para establecer este orden, utilizo el orden de muestras en el arbol producido por matriz_1.

```r
orden_matriz1 <- readRDS("data/orden_matriz1.Rds")
```

 \
Se necesita insalar y cargar los sgtes paquetes:

```r
library("ape")
library("tidyverse")
library("phytools")
library("dendextend")
```

## 1. Graficar una "resta de matrices"

\
Para graficar la diferencia entre matrices se recurre a restar ambas matrices y graficar los valores en la nueva matriz resultante.

```r
diferencia <- matriz1_dist_jaccard_darwin - matriz2_dist_jaccard_darwin
diferencia <- as.matrix(diferencia)
diferencia <- as.dist(diferencia)
diferencia <- as.matrix(diferencia)
diferencia <- diferencia[match(orden_matriz1, row.names(diferencia)),match(rev(orden_matriz1), row.names(diferencia))] # "orden" para filas y "reverso" para columnas
diferencia[1:6, 1:6]
```

La matriz ordenada tiene las columnas ordendas en sentido inverso a las filas para que el gráfico produzca una diagonal iniciando en la posicion 1:1

```r
min(diferencia, na.rm = TRUE)
max(diferencia, na.rm = TRUE)
```

```r
tibli_dif <- as_tibble(diferencia, rownames="A") %>%
  pivot_longer(-A, names_to = "B", values_to = "distances")
```

```r
head(tibli_dif)
```

```r
library(repr)
```

```r
options(repr.plot.width=8, repr.plot.height=7)

ggplot(tibli_dif, aes(x=fct_inorder(A), y=fct_inorder(B), fill=distances))+
  geom_tile()+
  scale_fill_gradient2(low = "#0000ff", high = "red", mid = "white", midpoint = 0.01, limit = c(-0.0051,0.03667), space = "Lab", name="difference")+
  theme(axis.text.x=element_blank(),
        axis.ticks.x=element_blank(),
        axis.text.y=element_blank(),
        axis.ticks.y=element_blank() 
        )

options(repr.plot.width=7, repr.plot.height=7)
```

Según los colores configurados en el gráfico se pueden identificar el grado de variacion entre las matrices. Aqui los valores azules indican variaciones cercanas a cero mientras que colores rojos indican mayores variaciones.

# 2. Tanglegram (dendextend)

\
Otra forma de comparar se hace graficando los arboles producidos por cada matriz de distancia. Estos árboles deben  estar previamente "enraizados" (lo hice manualmente en iTOL) y en algunos casos y en algunos árboles necesité rotar algún nodo para obtener la misma secuencia de clusters en todos los árboles y así comparar correcamente.

```r
darwin1_root <- read.tree("data/darwin1_rooted.txt")
darwin6_root <- read.tree("data/darwin6_rooted.txt")
```

```r
#calculando el árbol en DARwin, ninguna rama tiene valores negativos, esto sí sucede calculando desde R

which(darwin1_root$edge.length < 0)
which(darwin6_root$edge.length < 0)
```

```r
# por ser NJ, no cumple esta propiedad

is.ultrametric(darwin1_root)
is.ultrametric(darwin6_root)
```

```r
# Ya fueron enraizados manuamente en iTOL y vueltos a exportar en formato Newick

is.rooted(darwin1_root)
is.rooted(darwin6_root)
```

```r
darwin1_root <- force.ultrametric(darwin1_root, method = "extend")
darwin6_root <- force.ultrametric(darwin6_root, method = "extend")
```

```r
is.ultrametric(darwin1_root)
is.ultrametric(darwin6_root)
```

\
Se carga la informacion de texto y color para cada muestra en el árbol.

```r
labels_darwin1_root <- readRDS("data/labels_darwin1_root.Rds")
labels_darwin6_root <- readRDS("data/labels_darwin6_root.Rds")
color_darwin1_root <- readRDS("data/color_darwin1_root.Rds")
color_darwin6_root <- readRDS("data/color_darwin6_root.Rds")
```

\
Lo siguiente es la secuencia de pasos para editar ciertos atributos del arbol que seran necesarios para la comparación.

```r
darwin1_root %>%
  as.dendrogram() %>%
  set("labels" ,labels_darwin1_root) %>%
  set("labels_col", color_darwin1_root)  %>% # change color
  set("labels_cex", 0.5) %>%
  plot()
```

\
Verificada la correcta edicion de los "labels" y sus respectivos colores, procedemos a crear nuevos árboles que contengan la edición anteriormente realizada

```r
darwin1_root_tngl <- darwin1_root %>%
  as.dendrogram() %>%
  set("labels" ,labels_darwin1_root) %>%
  set("labels_col", color_darwin1_root)
```

```r
darwin6_root_tngl <- darwin6_root %>%
  as.dendrogram() %>%
  set("labels" ,labels_darwin6_root) %>%
  set("labels_col", color_darwin6_root)
```

```r
options(repr.plot.width=7, repr.plot.height=13)

tanglegram(darwin1_root_tngl, darwin6_root_tngl, lab.cex= 0.3, lwd=2, edge.lwd=1.5)  # 6x20 inch tambien es bueno

options(repr.plot.width=7, repr.plot.height=7)
```
